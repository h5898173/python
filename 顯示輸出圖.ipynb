```python
import matplotlib.pyplot as pt
import numpy as np
x=np.arange(0,360)
y=np.sin(x*np.pi/180.0)
pt.xlim(0,360)
pt.ylim(-1,2,1,2)
pt.plot(x,y)
pt.show()
```

```python
import matplotlib.pyplot as pt
import numpy as np
x=np.arange(0,360)
y=np.sin(x*np.pi/180.0)
z=np.cos(x*np.pi/180.0)
pt.xlim(0,360)
pt.ylim(-1,2,1,2)
pt.plot(x,y,label="sin")
pt.plot(x,z,label="cos")
pt.title("SIN and COS")
pt.xlabel("Degree")
pt.ylabel("Value")
pt.legend()
pt.show()
```

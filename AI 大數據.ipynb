```python
from sklearn import datasets,cluster,metrics
iris = datasets.load_iris()
iris_data =iris.data 
result=cluster.KMeans(n_clusters=3).fit(iris.data)
print(result.labels_)
silhouette_avg = metrics.silhouette_score(iris_data,result.labels_)
print(silhouette_avg)
```

```python
from sklearn import datasets,cluster,metrics
import matplotlib.pyplot as pt
iris=datasets.load_iris()
iris_data=iris.data

ks=range(2,11)
silhouette_avgs=[]
for n in ks:
    result=cluster.KMeans(n_clusters=n).fit(iris_data)
    silhouette_avg=metrics.silhouette_score(iris_data,result.labels_)
    silhouette_avgs.append(silhouette_avg)

pt.bar(ks,silhouette_avgs)
pt.show()
```

```python
from sklearn import datasets,svm
import matplotlib.pyplot as plt
digits=datasets.load_digits()
from sklearn.cross_validation import train_test_split
x_train,x_test,y_train,y_text,images_train,images_test = train_test_split(digits.data,digits.target,digits.images,test_size=0.25,random_state=66)

svc_model = svm.SVC(gamma=0.001,C=100,kernel='linear')
svc_model.fit(x_train,y_train)
predicted = svc_model.predict(x_test)
images_and_predictions= list(zip(images_test, predicted))

for index,(image,prediction) in enumerate(images_and_predictions[:5]):
    plt.subplot(1,5,index+1)
    plt.axis('off')
    plt.imshow(image,cmap=plt.cm.binary)
    plt.title(' Predicted:'+str(prediction))
plt.show()    
```
